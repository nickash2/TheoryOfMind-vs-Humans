```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('game_statistics.csv')
```

```python
stats
```

```python
# Plot results
N_RUNS = 10
fig, ax = plt.subplots(figsize=(10, 6))
x_pos = np.arange(len(stats.index))

ax.bar(
    x_pos, stats["Mean"], yerr=stats["Std"], align="center", alpha=0.7, capsize=10
)
ax.set_ylabel("Scores", fontsize=15)
ax.set_xlabel("Agents", fontsize=15)
ax.set_xticks(x_pos)
ax.set_xticklabels(stats.index)

# Set the x-axis labels to be the names of the agents
agent_names = [stats.iloc[i][0] for i in range(len(stats))]
ax.set_xticklabels(agent_names)

ax.set_title(
    f"Player Performance of {agent_names[0]} against {agent_names[1]} Over {N_RUNS} runs",
    fontsize=19
)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
# change the figure  text
plt.show()
```

```python
plt.savefig("performance_summary.png")
plt.close()
```
